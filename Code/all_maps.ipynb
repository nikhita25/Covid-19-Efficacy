```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
import folium 
import mapclassify
```

```python
Data=pd.read_csv('../Data/covid_data_merged')
```

```python
Data=Data.drop(columns=['Protection from Ancestral var - Severe Disease','Protection from Ancestral var - Infection','Protection from Beta var - Severe Disease','Protection from Beta var - Infection','Protection from Gamma var - Severe Disease','Protection from Gamma var - Infection'])
```

```python
#Data=Data.drop(labels= 'European Union', axis=0)
new_dt=Data[Data.Country != 'European Union']
```

```python
new_dt
```

```python
new_dt.groupby('Country')['Total_Vaccinations'].max();
```

```python
new_dt['Vaccine_Manufacturer'].unique()
```

```python
Oxford_AstraZeneca=new_dt[new_dt['Vaccine_Manufacturer']=='Oxford/AstraZeneca']
```

```python
maxi=pd.DataFrame(Oxford_AstraZeneca.groupby('Country')['Total_Vaccinations'].max())
```

```python
maxi.reset_index(inplace=True)
```

```python
# Read the geopandas dataset
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
tables=world.merge(maxi,how='left',left_on=['name'],right_on=['Country'])
```

```python
tables=tables.dropna(subset=['Total_Vaccinations'])
```

```python
# Create a map
my_map = folium.Map()
```

```python
folium.Choropleth(
    geo_data=tables,
    name='choropleth',
    data=tables,
    columns=['Country', 'Total_Vaccinations'],
    key_on='feature.properties.name',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Oxford/AstraZeneca Vaccine'
).add_to(my_map)
```

```python
my_map
```
